```python
import matplotlib.pyplot as plt
import json
```

```python
file = open("data/2024-OCT-NOV-CB-temp-precip.json", "r")
data = json.load(file)

dates = data["data"]["times"]
temps = data["data"]["temperature"]
precips = data["data"]["precipitation"]

# make a median temp for each 10 plots
median_temps = []
median_precips = []
for i in range(0, len(temps), 20):
	median_temps.append(sum(temps[i:i+10])/10)
	median_precips.append(sum(precips[i:i+10])/10)
```

```python
# plot line graph with temperature and precipitation

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
# make the x labels vertical
plt.xticks(rotation=90)
# dont show them all
plt.locator_params(axis='x', nbins=10)
ax1.set_ylabel('Temperature (C)', color=color)
ax1.plot(dates, temps, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Precipitation (mm)', color=color)
ax2.plot(dates, precips, color=color)
ax2.tick_params(axis='y', labelcolor=color)

ax3 = ax1.twinx()
color = 'tab:green'
ax3.set_ylabel('Median Temperature (C)', color=color)
ax3.plot(dates[::20], median_temps, color=color)
ax3.tick_params(axis='y', labelcolor=color)


fig.tight_layout()
plt.show()
```
